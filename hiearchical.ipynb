```python
import gensim.models
import gensim
```

```python
dictionary = gensim.corpora.Dictionary.load('larry.dict')
corpus = gensim.corpora.MmCorpus('larry.mm')

hdp = gensim.models.hdpmodel.HdpModel(corpus, dictionary)
print hdp
```

```python
hdp.print_topics(-1, topn=10)
```
